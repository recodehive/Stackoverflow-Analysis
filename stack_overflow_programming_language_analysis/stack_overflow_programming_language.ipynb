```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as p
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the csv file.
df = pd.read_csv("/content/TotalQuestions.csv",index_col=[0])

# Convert index to date type
if not isinstance(df.index, pd.DatetimeIndex):
    df.index = pd.to_datetime(df.index)
```

# **TIMELINE OF LANGUAGES USED BY STACK OVERFLOW USERS**

```python
# Group data by year
df_annual = df.resample('A').sum()

# Annual Line Chart
plt.figure(figsize=(14, 8))
for column in df_annual.columns:
    plt.plot(df_annual.index, df_annual[column], label=column)

plt.title('Timeline of the number of questions per category (2008-2024)')
plt.xlabel('Year')
plt.ylabel('Number of Questions')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# **Sudden Count Changes (Emerging Trends in Technology)**

```python
# Calculate the change in question counts for each programming language over time
df_change = df.diff()

# Plot the change in question counts for each programming language over time
plt.figure(figsize=(12, 8))
for column in df_change.columns:
    plt.plot(df_change.index, df_change[column], label=column)

plt.title('Change in Question Counts for Each Programming Language Over Time')
plt.xlabel('Month')
plt.ylabel('Change in Question Count')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# **Average Questions Per Month**

```python
# Aggregate the data by month and calculate the total number of questions for each month
monthly_total_questions = df.resample('M').sum().sum(axis=1)

# Group the data by month and extract the month names
monthly_total_questions_grouped = monthly_total_questions.groupby(monthly_total_questions.index.month).mean()

# Sort the average number of questions
sorted_months = monthly_total_questions_grouped.sort_values()

# Get the month with the lowest count
lowest_count_month = sorted_months.idxmin()

# Get the month with the highest count
highest_count_month = sorted_months.idxmax()

# Plot the total number of questions for each month over the entire period
plt.figure(figsize=(12, 8))

# Set colors for the lowest count month as red, highest count month as green, and default color for other months
colors = ['red' if month == lowest_count_month else 'green' if month == highest_count_month else 'skyblue' for month in monthly_total_questions_grouped.index]

monthly_total_questions_grouped.plot(kind='bar', color=colors)
plt.title('Average Number of Questions by Month (Entire Period)')
plt.xlabel('Month')
plt.ylabel('Average Number of Questions')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Add labels displaying the total number of questions on each bar
for bar in plt.gca().patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), int(bar.get_height()),
             ha='center', va='bottom', fontsize=9)

plt.show()
```

# **Average Questions Per year**

```python
# Aggregate the data by year and calculate the total number of questions for each year
yearly_total_questions = df.resample('Y').sum().sum(axis=1)

# Group the data by year and extract the year values
yearly_total_questions_grouped = yearly_total_questions.groupby(yearly_total_questions.index.year).mean()

# Sort the average number of questions
sorted_years = yearly_total_questions_grouped.sort_values()

# Get the year with the lowest count
lowest_count_year = sorted_years.idxmin()

# Get the year with the highest count
highest_count_year = sorted_years.idxmax()

# Plot the total number of questions for each year over the entire period
plt.figure(figsize=(12, 8))

# Set colors for the lowest count year as red, highest count year as green, and default color for other years
colors = ['red' if year == lowest_count_year else 'green' if year == highest_count_year else 'skyblue' for year in yearly_total_questions_grouped.index]

yearly_total_questions_grouped.plot(kind='bar', color=colors)
plt.title('Average Number of Questions by Year (Entire Period)')
plt.xlabel('Year')
plt.ylabel('Average Number of Questions')
plt.xticks(rotation=45)

# Add labels displaying the total number of questions on each bar
for bar in plt.gca().patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), int(bar.get_height()),
             ha='center', va='bottom', fontsize=9)

plt.show()
```

# **Total Questions By programming Languages**

```python
# Calculate the total number of questions for each language
total_questions_by_language = df.sum().sort_values(ascending=False)

# Plot the total number of questions by programming languages
plt.figure(figsize=(14, 8))

# Set colors for the plot
colors = sns.color_palette('coolwarm', len(total_questions_by_language))

# Plot the data
bars = plt.bar(total_questions_by_language.index, total_questions_by_language.values, color=colors)

# Add proper names and legend
plt.title('Total Number of Questions by Programming Languages', fontsize=16)
plt.xlabel('Programming Languages', fontsize=14)
plt.ylabel('Total Number of Questions', fontsize=14)
plt.xticks(rotation=45)

# Add labels displaying the total number of questions on each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', fontsize=9)

# Add legend
plt.legend(bars, total_questions_by_language.index, title='Languages', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

# **Top 10 Porgramming Languages of All time**

```python
# Calculate the total number of questions for each language
total_questions_by_language = df.sum().sort_values(ascending=False)

# Select the top 10 languages
top_10_languages = total_questions_by_language.head(10)

# Plot the total number of questions by programming languages
plt.figure(figsize=(14, 8))

# Set colors for the plot
colors = sns.color_palette('viridis', len(top_10_languages))

# Plot the data
bars = plt.bar(top_10_languages.index, top_10_languages.values, color=colors)

# Add proper names and legend
plt.title('Top 10 Programming Languages by Total Number of Questions', fontsize=16)
plt.xlabel('Programming Languages', fontsize=14)
plt.ylabel('Total Number of Questions', fontsize=14)
plt.xticks(rotation=45)

# Add labels displaying the total number of questions on each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom', fontsize=9)

plt.show()
```

# **Impact of different events on Questions timeline**

```python
# Extract data for top 10 languages by number of question asked
top_5_data = df.sum().sort_values(ascending=False).head(5)
top_5_languages = top_5_data.index.tolist()

# Filter the DataFrame to include only the top 10 languages
df_top_5 = df[top_5_languages]

# Plot individual temporal series for top 5 languages
plt.figure(figsize=(12, 8))

# Add vertical marker for historic events
event1_date = pd.to_datetime('2009-01-01')
event2_date = pd.to_datetime('2015-07-01')
event3_date = pd.to_datetime('2016-05-01')
event4_date = pd.to_datetime('2020-05-01')
event5_date = pd.to_datetime('2022-12-01')

for language in df_top_5.columns:
    plt.plot(df.index, df_top_5[language], label=language, marker='o', linestyle='-')

# Add vertical lines for historic events
plt.axvline(event1_date, color='green', linestyle='--', linewidth=2)
plt.axvline(event2_date, color='red', linestyle='--', linewidth=2)
plt.axvline(event3_date, color='orange', linestyle='--', linewidth=2)
plt.axvline(event4_date, color='grey', linestyle='--', linewidth=2)
plt.axvline(event5_date, color='brown', linestyle='--', linewidth=2)

# Define additional legend for markers
marker_legend = [plt.Line2D([0], [0], marker='|', color='green', markersize=10, linestyle='None'),
                 plt.Line2D([0], [0], marker='|', color='red', markersize=10, linestyle='None'),
                 plt.Line2D([0], [0], marker='|', color='orange', markersize=10, linestyle='None'),
                 plt.Line2D([0], [0], marker='|', color='grey', markersize=10, linestyle='None'),
                 plt.Line2D([0], [0], marker='|', color='brown', markersize=10, linestyle='None')]
marker_labels = ['Launch of 4G Phones','Introduction to Windows 10','Increase in Youtube usage', 'Post Covid-19 Work from Home', 'Launch of Chat-GPT']

plt.legend(loc='upper left')

# Add additional legend for markers
plt.legend(handles=marker_legend, labels=marker_labels, loc='upper left',bbox_to_anchor=(1.05, 1))

# Display plot
plt.title('Correlation with Events')
plt.xlabel('Date')
plt.ylabel('Number of Questions')
plt.xticks(rotation=45)
plt.show()
```

# **Number of Questions by days of the week**

```python
# Aggregate the data by day of the week and calculate the total number of questions for each day
daily_total_questions = df.groupby(df.index.dayofweek).sum().sum(axis=1)

# Create a custom color palette
colors = sns.color_palette("husl", len(daily_total_questions))

# Plot the total number of questions for each day of the week
plt.figure(figsize=(10, 6))
daily_total_questions.plot(kind='bar', color=colors)
plt.title('Total Number of Questions by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Questions')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.show()
```

# **Boxplot Distribution of top 10 Programming Languages**

# **Growth rate comparison of languages**

```python
# Calculate growth rates
first_value = df.iloc[0]
last_value = df.iloc[-1]
growth_rates = ((last_value / first_value) ** (1 / (len(df) - 1))) - 1

# Plotting
plt.figure(figsize=(12, 8))


growth_rates.plot(kind='bar', color=['blue', 'green', 'orange'])
plt.title('Growth Rate Comparison')
plt.xlabel('Programming Language')
plt.ylabel('Growth Rate')

plt.tight_layout()
plt.show()

print("\nGrowth Rates:")
print(growth_rates)
```

# **network plot for top 10 languages**

```python
import networkx as nx

# Extract top 10 languages by total questions
top_10_data = df.sum().sort_values(ascending=False).head(10)
top_10_languages = top_10_data.index.tolist()

# Filter the DataFrame to include only the top 10 languages
df_top_10 = df[top_10_languages]

# Calculate correlation matrix
corr_matrix = df_top_10.corr()

# Create a graph from the correlation matrix
G = nx.from_numpy_array(corr_matrix.values)

# Plotting the network
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # positions for all nodes

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=1500, node_color='skyblue', edgecolors='grey')

# Draw edges
nx.draw_networkx_edges(G, pos, edge_color='grey')

# Draw labels
nx.draw_networkx_labels(G, pos, labels={i: top_10_languages[i] for i in range(len(top_10_languages))}, font_size=10, font_weight='bold')

plt.title('Network Plot of Top 10 Programming Languages based on Correlation')
plt.show()
```
